# Test Pandas on extracted TM (KR5 One year 2015)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import dateutil
import glob
import os

DATA_PATH = r'/home/nicolargo/data/TM_EXTRACT_KR5'
all_files = glob.glob(os.path.join(DATA_PATH, '*.tsv'))

def mydateparser(d, dayfirst=True):
    return dateutil.parser.parse(d.replace('_', 'T'), dayfirst=True)
```

```python
assert len(all_files) == 53
```

```python
%matplotlib inline
```

```python
tm = pd.concat((pd.read_csv(f, sep='\t', date_parser=mydateparser, comment='#',  index_col=0, parse_dates=True, infer_datetime_format=True, error_bad_lines=False, squeeze=True) for f in all_files))
```

```python
ohlc_list = tm.TR0160K.fillna(method='pad').resample('W').ohlc()
```
