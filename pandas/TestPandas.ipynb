# Init

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import dateutil
def mydateparser(d, dayfirst=True):
    return dateutil.parser.parse(d.replace('_', 'T'), dayfirst=True)
```

```python
%matplotlib inline
```

```python
input_file = './input.tsv'
i = pd.read_csv(input_file, sep='\t', date_parser=mydateparser, comment='#',  index_col=0, parse_dates=True, infer_datetime_format=True, error_bad_lines=False, squeeze=True)
```
